### Box plot
เพื่อดูการกระจายของข้อมูล

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
iris_data = pd.read_csv('C:\\Users\\MyCom\\Documents\\DataViz\\iris.data' ,header=None)
iris_data.head()
```

- แบ่งข้อมูล ออกเป็น 3 classes คือ Setosa,Versicolor และ virginica

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
setosa_np = iris_data.iloc[:50,:-1].to_numpy()
versicolor_np = iris_data.iloc[50:100,:-1].to_numpy()
virginica_np = iris_data.iloc[100:,:-1].to_numpy()
```

##### Excemple Box plot

```python
fig1, ax1 = plt.subplots()
ax1.set_title('Basic Plot')
ax1.boxplot(setosa_np)
```

- แบ่ง setosa_np เป็น 4 feture เพื่อดูการกระจายของข้อมูล

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = [10,10]
plt.subplot(1,4,1)
plt.boxplot(setosa_np[:,0])
plt.show
plt.subplot(1,4,2)
plt.boxplot(setosa_np[:,1])
plt.show
plt.subplot(1,4,3)
plt.boxplot(setosa_np[:,2])
plt.show
plt.subplot(1,4,4)
plt.boxplot(setosa_np[:,3])
plt.show
```

### Bubble Chart
- คือ การใช้ขนาดของวงกลมแสดงค่าของ Data

```python
plt.scatter(setosa_np[:,0], setosa_np[:,1], s=3000*setosa_np[:,2], c=setosa_np[:,0], cmap="pink", alpha=0.4, edgecolors="grey", linewidth=2)
```

```python

plt.scatter(setosa_np[:,0], setosa_np[:,1], s=2000*setosa_np[:,2], c=setosa_np[:,0], cmap="bone", alpha=0.4, edgecolors="blue", linewidth=2,marker = '*')
```

```python
plt.scatter(virginica_np[:,3],virginica_np[:,0],s=2000*virginica_np[:,1], c = virginica_np[:,2], cmap="pink",alpha=0.8, edgecolors="grey", linewidth=2)
# Add titles (main and on axis)
plt.xlabel("virginica is Features 4") ##ชื่อแกน X
plt.ylabel("Virginica is Features 1") ##ชื่อแกนY
plt.title("Virginica bubble plot") ##ชื่อ Bubble chart
plt.show()
```

### Map
- การ Plot Map การสำรวจภาวะเศรษฐกิจและสังคมของครัวเรือน ปี 2549 โดยใช้แผนที่ประเทศไทย

```python
import folium #ติดตั้ง pip install folium
```

```python
Debt = 'C:\\Users\\MyCom\\Documents\\DataViz\\Data.xls'
```

```python
Debt_df = pd.read_excel(Debt,skiprows=3)
```

- เลือกเฉพาะคอลัมน์ที่สนใจ

```python
Debt_df[['Region and province', 2549]]
```

```python
Debt_df_debt = Debt_df[['Region and province', 2549]]
```

```python
Debt_df_debt.drop([0,1,2,3,8,31,49,70,85,86,87,88],axis =0,inplace=True)
```

```python
Debt_df_debt
```

```python
data_to_plot = Debt_df_debt
```

```python
data_to_plot
```

- Download ข้อมูลเชิงภูมิศาสตร์ของประเทศไทย จาก thai.json
- วน Loop ลบช่องว่างข้างหน้าชื่อออกเพื่อที่จะสามารถทำให้ไฟล์ 2 ไฟล์นั้นสามารถเชื่อมกันได้

```python
country_geo = 'thai.json'
```

```python
new_col = list()
for d in data_to_plot['Region and province']:
     new_col.append(d.lstrip())
data_to_plot['Region and province'] = new_col
```

- กำหนดพิกัด ลองติจูดและละติจูดของประเทศไทย

```python
map = folium.Map(location=[10,100],  zoom_start =5)
map
```

- ลงสีของข้อมูลที่มีค่าของจำนวนหนี้สินเฉลี่ยต่อครัวเรือน

- ถ้าจังหวัดไหนที่มีจำนวนหนี้สินเฉลี่ยต่อครัวเรือนมีค่ามากจะเป็นสีม่วงเข้ม

- จังหวัดไหนที่มีจำนวนหนี้สินเฉลี่ยต่อครัวเรือนน้อยจะเป็นสีเหลืองตามแถบ

```python
map.choropleth(geo_data=country_geo,data=data_to_plot,
                          columns=['Region and province',2549],
                          key_on='feature.properties.name',
                          fill_color='YlGnBu',fill_opacity=0.7,line_opacity=0.2,
                          legend_name='Average Debt')
```

```python
map.save('plot_data.html')
```
