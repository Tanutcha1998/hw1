### ตรวจสอบ missing

````python
Covid.isnull().any()
````

### แทนค่า missing ด้วย "unknown" 

````python
Covid['Province/State'] = Covid['Province/State'].fillna('unknown')
````

````python
Covid.isnull().any()
````

````python
Covid
````

### จัดการตารางที่ต้องการ
สร้าง columns >> 'Date_new' โดยเอาปีและเวลาออก

````python
D_T = []
for i in range(Covid.shape[0]):
    A = Covid.iloc[i]['Date'].split(' ')
    A = A[0].replace('/2020', '')
    D_T.append(A)
D_T
````

````python
Covid['Date_new'] = D_T
````

ลบ column>>'Date'และ'Last Update' ออก

````python
del Covid['Date']
del Covid['Last Update']
````

````python
Covid
````

````python
Main_China = Main_China.reset_index(drop=True)
````

````python
Main_China.mean()
````

### สร้างกราฟ

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
x = Covid.groupby(['Date_new'], as_index=False).sum()['Date_new']
y1 = Covid.groupby(['Date_new'], as_index=False).sum()['Confirmed']
y2 = Covid.groupby(['Date_new'], as_index=False).sum()['Deaths']
y3 = Covid.groupby(['Date_new'], as_index=False).sum()['Recovered']
plt.figure(figsize=[10,5],dpi=80)
plt.gca(facecolor='oldlace')
plt.plot(x,y1,'pink',label = "Confirmed",marker='o')
plt.plot(x,y2,'skyblue',label = "Deaths",marker='o')
plt.plot(x,y3,'lime',label = "Recovered",marker='o')
plt.legend()
plt.xticks(rotation = 90)
plt.xlabel('วันที่',fontname='Tahoma')
plt.ylabel('จำนวน(คน)',fontname='Tahoma')
plt.title('กราฟแสดงจำนวนผู้ติดเชื้อ Covid-19 เดือน ม.ค.- ก.พ. 2563',fontname='Tahoma',fontsize=16)
````

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
x = Covid[Covid['Country']=='Thailand']['Date_new']
y1 = Covid[Covid['Country']=='Thailand']['Confirmed']
y2 = Covid[Covid['Country']=='Thailand']['Deaths']
y3 = Covid[Covid['Country']=='Thailand']['Recovered']
plt.figure(figsize=[10,5],dpi=80)
ax = plt.gca(facecolor='#EFEEFE')
plt.fill_between( x, y1, color="skyblue",alpha=0.2,label = "Confirmed")
plt.plot(x, y1, color="Slateblue", alpha=0.6)
plt.fill_between( x, y2, color="violet",alpha=0.2,label = "Deaths")
plt.plot(x, y2, color="plum", alpha=0.6)
plt.fill_between( x, y3, color="aquamarine",alpha=0.2,label = "Recovered")
plt.plot(x, y3, color="turquoise", alpha=0.6)
plt.legend()
plt.xticks(rotation = 90)
ax.yaxis.set_ticks_position('right')
plt.xlabel('วันที่',fontname='Tahoma')
plt.ylabel('จำนวน(คน)',fontname='Tahoma')
plt.title('กราฟแสดงจำนวนผู้ติดเชื้อ Covid-19 ในประเทศไทย เดือน ม.ค.- ก.พ. 2563',fontname='Tahoma',fontsize=12)
````
