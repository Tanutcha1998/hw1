ข้อ1. สร้าง function ชื่อ smartDict ซึ่งทำหน้าที่:
- function รับ input เป็น list 2 ตัวที่มีขนาดเท่ากัน โดย list แรกเป็น list ของตัวเลข list ที่สองเป็น list ของ text
- ถ้าค่าใน list แรก เป็น เลขคี่ ให้ค่าใน dict ที่เป็น output มี key เป็นค่าจาก list ตัวที่ 1 และ value จาก list ตัวที่ 2
- ถ้าค่าใน list แรก เป็น เลขคู่ ให้ค่าใน dict ที่เป็น output มี key เป็นค่าจาก list ตัวที่ 2 และ value จาก list ตัวที่ 1
- output ของ function คือ dict ที่สร้างขึ้น

เช่น
output1 = smartDict([ 1 , 2 , 4 ] , [ 'apple' , 'banana' , 'orange' ]  )

จะได้ output1 เป็น { 1:'apple' , 'banana':2 , 3:'orange' }

```python
def smartDict(listnumber,listtext):
    retdict = {}
    for i in range(len(listnumber)):
        if listnumber[i]%2 == 1 :
            retdict[listnumber[i]] = listtext[i]
        else :
            retdict[listtext[i]] = listnumber[i]
    return retdict
```

```python
smartDict([ 1 , 2 , 4 ] , [ 'apple' , 'banana' , 'orange' ] )
```

```python
smartDict([ 7 , 1 , 4 , 6 ] , [ 'A' , 'B' , 'E' ,'Zzz' ] )
```

ข้อ2.2 จงหา sequence ที่มีขนาด 3 และ 4 ที่เกิดขึ้นซ้ำมากที่สุด

```python
def populate_n_char(seq,n):
    str_list = list()
    for i in range(len(seq)-(n-1)) :
        str_list.append(seq[i:i+n])
    return str_list
```

```python
def seq_n(seq,n):
    list_all_n_char = list()
    for row in range(DNAseq.shape[0]):
        list_all_n_char = list_all_n_char + populate_n_char(DNAseq.iloc[row][0],n)
    unique_n_char = list(set(list_all_n_char))
    dict_n_char = {}
    for sub_str in unique_n_char :
        num = 0
        for s in list_all_n_char :
            if sub_str == s :
                num += 1
        dict_n_char[sub_str] = num
    maxValue = -100
    maxKey = None
    for k in dict_n_char.keys():
        if dict_n_char[k] > maxValue:
            maxValue = dict_n_char[k]
            maxKey = k
    print(f'sequence ที่มีขนาด {n} ที่เกิดขึ้นซ้ำมากที่สุด คือ {maxKey} จำนวน {maxValue}')
```
