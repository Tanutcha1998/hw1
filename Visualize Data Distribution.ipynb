## Scatter Plot
ดูการกระจายของข้อมูล

import pandas เข้ามาและตั้งชื่อใหม่เป็น pd

อ่านไฟล์ iris.data กำหนดหัวตารางเป็น None แล้วตั้งชื่อตารางเป็น irisdata

```python
import pandas as pd
irisdata = pd.read_csv('C:\\Users\\MyCom\\Documents\\DataViz\\iris.data',header = None)
irisdata
```

import numpy และเปลี่ยนชื่อเป็น np เพื่อที่จะใช้คำสั่ง iloc ได้

```python
import numpy as np
```

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()
setosa_np
```

การวาด scatter plot อย่างง่าย ใช้ plt.scatter( x , y )

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1])
```

s แทน ขนาด marker(จุด)

c แทน สีของจุด

s = 10*setosa_np[:,2] คือ กำหนดให้ขนาดของจุดเท่ากับสิบเท่าของข้อมูลในหลักที่3

marker แทน รูปแบบของ marker

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s = 10*setosa_np[:,2])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s = 10*setosa_np[:,2],c='c')
```

```python
print(f'max = {max(setosa_np[:,3])}, min = {min(setosa_np[:,3])}')
```

การกำหนดสีแต่ละช่วงค่าข้อมูล

- ถ้าค่าในหลักที่4 น้อยกว่า 0.15 ให้เป็นสีน้ำเงิน
- ถ้าค่าในหลักที่4 น้อยกว่า 0.30 ให้เป็นสีเขียว
- ถ้าค่าในหลักที่4 น้อยกว่า 0.45 ให้เป็นสีเหลือง
- ถ้าค่าในหลักที่4 มากกว่าหรือเท่ากับ 0.45 ให้เป็นสีแดง

```python
colors = []
for i in setosa_np[:,3]:
    if i < 0.15:
        colors.append('b')
    elif i < 0.30:
        colors.append('g')
    elif i < 0.45:
        colors.append('y')
    else:
        colors.append('r')
colors
```

แทนค่า c เป็นสี colors จากที่เราสร้าง

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s = 30*setosa_np[:,2],c=colors)
```

กำหนดค่าสีตามค่าข้อมูลในหลักที่4

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s = 30*setosa_np[:,2],c=setosa_np[:,3])
```

cmap แทน สี colormapที่เรากำหนด

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s = 30*setosa_np[:,2],c=setosa_np[:,3],cmap='YlOrBr')
```

การใส่ค่าสีโดยใช้ผสม [R,G,B] มีค่าตั้งแต่ 0-1 โดยใช้ np.array([[R,G,B]])

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s = 30*setosa_np[:,2],c=np.array([[0.5,0.5,1]]))
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s = 30*setosa_np[:,2],c=setosa_np[:,3],cmap='YlOrBr',marker="*")
```

### visualize 2 data

```python
virginica_np = irisdata.iloc[100:,:-1].to_numpy()
```
